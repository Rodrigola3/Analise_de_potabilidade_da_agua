## Ãrvore de DecisÃ£o - Decision Trees

### IntroduÃ§Ã£o do Algoritmo 

Os algoritmos de aprendizagem baseados em Ã¡rvores de decisÃ£o sÃ£o considerados um dos melhores e mais utilizados mÃ©todos de aprendizagem supervisionada. Os mÃ©todos baseados em Ã¡rvores nos dÃ£o modelos preditivos de alta precisÃ£o, estabilidade e facilidade de interpretaÃ§Ã£o. Ao contrÃ¡rio dos modelos lineares, eles mapeiam muito bem relaÃ§Ãµes nÃ£o-lineares. E podem ser adaptados para resolver vÃ¡rios tipos de problema (classificaÃ§Ã£o ou regressÃ£o).

Como nosso projeto se trata de um problema de ClassificaÃ§Ã£o entre Ãgua PotÃ¡vel e Ãgua NÃ£o-PotÃ¡vel, o algoritmo da Ãrvore de decisÃ£o Ã© um bom mÃ©todo a ser empregado!

### Mas afinal, o que Ã© uma Ãrvore de DecisÃ£o e como ela funciona?

Ãrvore de decisÃ£o Ã© um tipo de algoritmo de aprendizagem supervisionada (com uma variÃ¡vel alvo prÃ©-definida), muito utilizada em problemas de classificaÃ§Ã£o. Ele funciona para ambas as variÃ¡veis categÃ³ricas e contÃ­nuas de entrada e de saÃ­da. Na Ã¡rvore de decisÃ£o, dividimos a populaÃ§Ã£o ou amostra em dois ou mais conjuntos homogÃªneos (ou sub-populaÃ§Ãµes) com base nos divisores/diferenciadores mais significativos das variÃ¡veis de entrada.

### Como construir uma Ã¡rvore de decisÃ£o?

Como qualquer modelo de *machine learning*, Ã© necessÃ¡rio entender um pouco do processo matemÃ¡rico que o embasa. A Ã¡rvore de decisÃ£o que serÃ¡ construida nesse modelo serÃ¡ utilizada para classificaÃ§Ã£o. Para esse caso, hÃ¡ 2 critÃ©rios de divisÃ£o mais conhecidos: *Entropia* e *Ãndice de Gini*.

**A) Entropia**

A Entropia Ã© mais conhecida por medir o "grau de desordem" de um sistema. Nas Ã¡rvores de decisÃ£o, a entropia mede a falta de *homogeneidade* de uma amostra, isto Ã©, mede a impureza dos dados relacionados Ã  sua classificaÃ§Ã£o. Em outras palavras, a entropia faz o controle de como a Ãrvore decide a divisÃ£o dos dados.

O cÃ¡lculo da entropia ($S$) pode ser feito conforme a equaÃ§Ã£o:

$$Entropia (S) = - \sum_{i=1}^n p_i \cdot \log_{2}p_i$$

em que:
- $S$ Ã© um conjunto de dados com $n$ classes diferentes
- $p_i$ Ã© a fraÃ§Ã£o de dados de $S$ que pertencem Ã  classe $i$

A partir dos cÃ¡lculos de Entropia, calcula-se o Ganho de informaÃ§Ã£o que, em outras palavras, pode ser entendido como o grau de importÃ¢ncia de um dado atributo. Assim, o ganho de um dado atributo $A$ de um conjunto de dados $S$ Ã© dado por:

$$Ganho(S, A) = Entropia(S) - \sum_{v\in Valores(A)}\frac{\vert S_v \vert}{\vert S \vert}Entropia(S_v)$$

em que 
- $Valores(A)$ representa os possÃ­veis valores de $A$
- $v$ Ã© um elemento de $Valores(A)$
- $S_v$ Ã© o subconjunto de $S$ quando $x=A$

**B) Ãndice de Gini**

O Ãndice de Gini Ã© uma expressÃ£o matemÃ¡tica que calcula a impureza de um nÃ³ (grau de heterogeneidade dos dados) e Ã© determinado por:

$$Gini = 1 - \sum_{i=1}^n p_i^2$$

em que:
- $p_i$ Ã© a fraÃ§Ã£o de cada classe em cada nÃ³
- $n$ Ã© o nÃºmero de classes

# Modelo 02

## Floresta AleatÃ³ria - Random Forest

### IntroduÃ§Ã£o


Este nome explica muito bem o funcionamento do algoritmo, que irÃ¡ criar muitas Ã¡rvores de decisÃ£o, de maneira aleatÃ³ria, formando o que podemos enxergar como uma floresta, onde cada Ã¡rvore serÃ¡ utilizada na escolha do resultado final.

### Como funciona o algoritmo RandomForest

Como jÃ¡ mencionamos, no algoritmo RandomForest serÃ£o criadas vÃ¡rias Ã¡rvores de decisÃ£o, felizmente acabamos de trabalhar as principais ideias de Ãrvores de DecisÃ£o neste Notebook no modelo anterior.

**No caso do Random Forest, ele gera vÃ¡rios decision trees, cada um com suas particularidades e combinada o resultado da classificaÃ§Ã£o de todos eles. Essa combinaÃ§Ã£o de modelos, torna ele um algoritmo muito mais poderoso do que o Decision Tree.**

### SeleÃ§Ã£o de amostras

Diferentemente do que acontece na criaÃ§Ã£o de uma Ã¡rvore de decisÃ£o simples, ao utilizar o RandomForest, o primeiro passo executado pelo algoritmo serÃ¡ selecionar aleatoriamente algumas amostras dos dados de treino, e nÃ£o a sua totalidade. Nesta etapa Ã© utilizado o bootstrap, que Ã© um mÃ©todo de reamostragem onde as amostras selecionadas podem ser repetidas na seleÃ§Ã£o. Com esta primeira seleÃ§Ã£o de amostras serÃ¡ construÃ­da a primeira Ã¡rvore de decisÃ£o.



Link para imagem: http://carlosbaia.com/2016/12/24/decision-tree-e-random-forest/

### SeleÃ§Ã£o das variÃ¡veis para cada nÃ³

Conforme vimos na construÃ§Ã£o de uma Ã¡rvore de decisÃ£o, para comeÃ§ar Ã© preciso definir o primeiro nÃ³ da Ã¡rvore (nÃ³ raiz), que serÃ¡ a primeira condiÃ§Ã£o verificada, dando origem aos dois primeiros ramos. Utilizando o **algoritmo de entropia ou o Ã­ndice Gini, serÃ¡ escolhida a melhor variÃ¡vel para compor o nÃ³ raiz, variando de acordo com o mÃ©todo utilizado.

**No RandomForest a definiÃ§Ã£o desta variÃ¡vel nÃ£o acontece com base em todas as variÃ¡veis disponÃ­veis. O algoritmo irÃ¡ escolher de maneira aleatÃ³ria (random) duas ou mais variÃ¡veis**, e entÃ£o realizar os cÃ¡lculos com base nas amostras selecionadas, para definir qual dessas variÃ¡veis serÃ¡ utilizada no primeiro nÃ³. Para escolha da variÃ¡vel do prÃ³ximo nÃ³, novamente serÃ£o escolhidas duas (ou mais) variÃ¡veis, excluindo as jÃ¡ selecionadas anteriormente, e o processo de escolha se repetirÃ¡. Desta forma a Ã¡rvore serÃ¡ construÃ­da atÃ© o Ãºltimo nÃ³. A quantidade de variÃ¡veis a serem escolhidas pode ser definida na criaÃ§Ã£o do modelo.

Na construÃ§Ã£o da prÃ³xima Ã¡rvore, os dois processos anteriores se repetirÃ£o, levando a criaÃ§Ã£o de uma nova Ã¡rvore. Provavelmente essa Ã¡rvore serÃ¡ diferente da primeira, pois tanto na seleÃ§Ã£o das amostras, quanto na seleÃ§Ã£o das variÃ¡veis, o processo acontece de maneira aleatÃ³ria.

Podemos construir quantas Ã¡rvores quisermos, sendo que quanto mais Ã¡rvores criadas, melhor serÃ£o os resultados do modelo, atÃ© determinado ponto, onde uma nova Ã¡rvore nÃ£o conseguirÃ¡ levar a uma melhora significativa no desempenho do modelo. Ã‰ importante lembrarmos que quanto mais Ã¡rvores forem criadas, maior serÃ¡ o tempo de criaÃ§Ã£o do modelo.

### Vantagens do Random Forest:
1. O algoritmo nÃ£o Ã© enviesado, uma vez que hÃ¡ mÃºltiplas Ã¡rvores, em que cada uma Ã© treinada com um subconjunto dos dados. Basicamente, o algoritmo se apoia no *"poder da multidÃ£o"* para reduzir viÃ©ses individuais.
2. O algoritmo Ã© muito estÃ¡vel. Mesmo que uma nova linha de dado seja incrementada ao *dataset*, o algoritmo nÃ£o Ã© fortemente impactado, uma vez que o novo dado afetarÃ¡ uma Ã¡rvore, mas dificilmente afetarÃ¡ todas elas e, consequentemente, Ã  floresta.
3. O algoritmo funciona bem tanto para variÃ¡veis numÃ©ricas quanto categÃ³ricas.
4. O algoritmo tambÃ©m funciona quando hÃ¡ dados faltantes.

### Desvantagens do Random Forest:
1. A maior desvantagem do algoritmo se dÃ¡ em sua complexidade. Ele pode demandar altos recursos computacionais, a depender do nÃºmero de Ã¡rvores de decisÃ£o desejados ou atÃ© mesmo do tamanho da base de dados a ser trabalhada.
2. Ã‰ uma tÃ©cnica que demanda muito mais tempo, se comparada a outros algoritmos.
